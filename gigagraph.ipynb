```python
import matplotlib.pyplot as plt
import pandas as pd
import os
import subprocess

def run_shortest_way_program():
    relative_path = os.path.join('build', 'Debug', 'shortest_way.exe')

    if not os.path.exists(relative_path):
        print(f"error: hull_graph_shortest_way program '{relative_path}' was not found")
    
    try:
        final_vertex_index = '99'
        result = subprocess.run([relative_path, final_vertex_index], capture_output=True, text=True)
        print(result.stdout)
    except Exception as e:
        return f"hull_graph_shortest_way execution failed: {e}"

print(run_shortest_way_program())
```

```python
def load_points(filename):
    data = pd.read_csv(filename)
    return data['x'].values, data['y'].values

def plot_path(x, y):
    assert len(x) > 0
    assert len(x) == len(y)

    plt.figure(figsize=(8, 8))
    
    for i in range(len(x) - 1):
        plt.plot([x[i], x[i + 1]], [y[i], y[i + 1]], markersize=0.1, color='b', mfc='r', mec='b')
    
    for i in range(len(x) - 1):
        #plt.arrow(x[i], y[i], x[i + 1] - x[i], y[i + 1] - y[i])
        
        mid_x = (x[i] + x[i + 1]) / 2
        mid_y = (y[i] + y[i + 1]) / 2
        #plt.text(mid_x, mid_y, str(i), fontsize=8, ha='center', color='blue')
    
    plt.scatter(x, y, color='red', zorder=5)
    
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Shortest way path')
    plt.grid(True)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()

filename = 'shortest_way.csv' 

x, y = load_points(filename)

plot_path(x, y)
```
